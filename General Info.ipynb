```python
import pandas as pd

df = pd.read_excel(r"Cleaned_Tagoloan\20210101_20220101 Report\20210101_20220101 Report.xlsx", index_col=0)
df.head()
```

```python
df = df.drop(["Hour", "Rain Gauge 0", "Rain Gauge 1", "Water Level"], axis=1)

#Change Timestamp to Datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

#Add month and Year column
df['Month'] = df['Timestamp'].dt.month_name()
df['Year'] = df['Timestamp'].dt.d

df.head()
```

```python
#Create an array of dataframes where they are grouped by month and year
verbose_month = [i for _,i in df.groupby(by=['Month', 'Year'])]

#pass every item in the arry into the describe function of pandas and include name 
summative_month = [(f"{i.iloc[0,:]['Month']} {i.iloc[0,:]['Year']}", 
                    int(f"{i.reset_index(drop=True).loc[0,'Year']}{str(i.reset_index(drop=True).loc[0,'Timestamp'].month).zfill(2)}"), #outputs the first timestamp of the month withformat <year><month>
                    i.drop(["Year", "Month"], axis=1).describe()) for i in verbose_month]

#sort the list by order of month and year
summative_month.sort(key = lambda x: x[1])
```

```python
#Save to a spreadsheet
with pd.ExcelWriter("Summative Report.xlsx") as writer:
    for report in summative_month:
        name    = report[0]
        df      = report[2]

        df.to_excel(writer, sheet_name=name)
```
