```python
import pandas as pd

df = pd.read_csv(r"Edited_Data\Tagoloan\rg1.csv")
df
```

```python
df = df.drop_duplicates(subset="Timestamp", keep="first")
df
```

```python
df = df.iloc[:,4:]
df
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df = df.sort_values(by='Timestamp').reset_index(drop=True)
df
```

```python
df['Day Index'] = df['Timestamp'].apply(lambda x: x.strftime('%m-%d-%Y'))
df.drop(columns=df.columns[[1,2,4]], inplace=True)
```

```python
df
```

```python
day_df = df.groupby(by="Day Index").sum()
day_df
```

```python
day_df.index= pd.to_datetime(day_df.index)
day_df
```

```python
import numpy as np
day_df.loc[np.logical_and(day_df.index.month == 4, day_df.index.year == 2021)]
```

```python
df = df.drop(columns=df.columns[[2]])
df
```

```python
mo_and_year_df = df.copy()
mo_and_year_df['Month and Year'] = df['Timestamp'].apply(lambda x: x.strftime('%m-%Y'))
```

```python
summative_rainfall_by_month = mo_and_year_df.drop(columns=['Timestamp']).groupby(by='Month and Year').sum()
summative_rainfall_by_month
```

```python
average_rainfall_by_month = mo_and_year_df.drop(columns=['Timestamp']).groupby(by='Month and Year').mean()
average_rainfall_by_month
```
